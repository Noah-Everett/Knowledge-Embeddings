Nearest neighbors to the synthetic 'basis' vectors.

```python
IN_FILE = 'hepex-embeddings.npz'
```

```python
import pandas as pd
papers = pd.read_pickle(IN_FILE)
papers.head()
```

```python
import numpy as np
embeddings = papers['embedding'].to_list()
embeddings = np.vstack(embeddings)
```

```python
import cluster_utils as cu

items = papers.to_dict("records")

# Example: nearest neighbors to the first paper vector
pairs = cu.find_closest_point(
    embeddings,
    items,
    idx=0,
    top_k=10,
    require_different_group=False,
)

cu.point_pairs_to_dataframe(pairs, items)
```
